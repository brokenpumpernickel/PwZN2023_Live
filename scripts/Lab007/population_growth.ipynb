```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

$$\frac{dN}{dt}=r N \left(1-\frac{N}{K}\right)$$

```python
odeint
```

```python
def growth_model(y, t, r, K):
    return r * y * (1 - y / K)
```

```python
t = np.linspace(0, 10, 100)
y0 = 1
r = 1
K = 10

y = odeint(growth_model, y0, t, args=(r, K))
```

```python
plt.figure(figsize=(7, 4))
plt.plot(t, y, 'C1--', lw=2)
plt.plot(t, y / 2, 'C2--', lw=2)
plt.xlabel('t')
plt.ylabel('N')
plt.title('Logistic growth model')
plt.xlim(0, 10)
plt.ylim(0, 11)
plt.grid(linestyle='--')
```

```python
plt.figure(figsize=(10, 6))

ax = plt.subplot(221)
ax.plot(t, y, 'C1--', lw=2)
ax.set_title('a')

ax = plt.subplot(222)
ax.plot(t, y, 'C2--', lw=2)
ax.set_title('b')

ax = plt.subplot(223)
ax.plot(t, y, 'C3--', lw=2)
ax.set_title('c')

ax = plt.subplot(224)
ax.plot(t, y, 'C3--', lw=2)
ax.set_title('d')
```

```python
fig = plt.figure(figsize=(7, 4))
fig.set_tight_layout(True)

ax = plt.subplot()
line, = ax.plot([], [], 'C1--', lw=2)
ax.set_xlim(0, 10)
ax.set_ylim(0, 11)

plt.close()

def animate(i):
    line.set_data(t[:i], y[:i])
    
    return line,

anim = FuncAnimation(fig, animate, frames=len(t), interval=40, blit=True)
HTML(anim.to_jshtml())
```

```python
fig = plt.figure(figsize=(7, 4))
fig.set_tight_layout(True)

ax = plt.subplot()
line_lines, = ax.plot([], [], 'C1--', lw=2)
line_point, = ax.plot([], [], 'C2o', lw=2)
ax.set_xlim(0, 10)
ax.set_ylim(0, 11)

plt.close()

def animate(i):
    line_lines.set_data(t[:i], y[:i])
    line_point.set_data([t[i]], [y[i]])
    
    return line_lines, line_point

anim = FuncAnimation(fig, animate, frames=len(t), interval=40, blit=True)
HTML(anim.to_jshtml())
```
