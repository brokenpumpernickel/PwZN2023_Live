```python
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import row, column, gridplot, layout
from bokeh.models import Slider, Div
from bokeh.util.hex import hexbin
from bokeh.transform import linear_cmap
from bokeh.palettes import all_palettes

output_notebook()
```

```python
ts1 = np.linspace(0, 10, 1000)
ys1 = np.sin(ts1)
ys2 = np.cos(ts1)
ys3 = ys1 + ys2
```

```python
fig = figure(
    sizing_mode='stretch_width',
    aspect_ratio=3,
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig.grid.grid_line_dash = [6, 4]
fig.toolbar.logo = None
fig.toolbar.autohide = True

fig.line(ts1, ys1, color='red', legend_label='sin', line_width=3)
fig.line(ts1, ys2, color='pink', legend_label='cos', line_width=3)
fig.line(ts1, ys3, color='orange', legend_label='sin + cos', line_width=3)
fig.circle(ts1[::20], ys3[::20], color='orange', legend_label='sin + cos', radius=0.05, alpha=0.5)

show(fig)
```

```python
fig1 = figure(
    sizing_mode='stretch_width',
    aspect_ratio=1,
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig1.grid.grid_line_dash = [6, 4]
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.line(ts1, ys1, color='red', legend_label='sin', line_width=3)

fig2 = figure(
    sizing_mode='stretch_width',
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig2.grid.grid_line_dash = [6, 4]
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.line(ts1, ys2, color='green', legend_label='cos', line_width=3)

fig3 = figure(
    sizing_mode='stretch_width',
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig3.grid.grid_line_dash = [6, 4]
fig3.toolbar.logo = None
fig3.toolbar.autohide = True
fig3.line(ts1, ys3, color='blue', legend_label='sin + cos', line_width=3)

fig4 = figure(
    sizing_mode='stretch_width',
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig4.grid.grid_line_dash = [6, 4]
fig4.toolbar.logo = None
fig4.toolbar.autohide = True
fig4.circle(ts1[::20], ys3[::20], color='orange', legend_label='sin + cos', radius=0.05, alpha=0.5)

show(row(fig1, fig2, fig3, fig4, sizing_mode='stretch_width'))
```

```python
fig1 = figure(
    aspect_ratio=1,
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig1.grid.grid_line_dash = [6, 4]
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.line(ts1, ys1, color='red', legend_label='sin', line_width=3)

fig2 = figure(
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig2.grid.grid_line_dash = [6, 4]
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.line(ts1, ys2, color='green', legend_label='cos', line_width=3)

fig3 = figure(
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig3.grid.grid_line_dash = [6, 4]
fig3.toolbar.logo = None
fig3.toolbar.autohide = True
fig3.line(ts1, ys3, color='blue', legend_label='sin + cos', line_width=3)

fig4 = figure(
    aspect_ratio=1,    
    title='Mj Genialny Wykres',
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig4.grid.grid_line_dash = [6, 4]
fig4.toolbar.logo = None
fig4.toolbar.autohide = True
fig4.circle(ts1[::20], ys3[::20], color='orange', legend_label='sin + cos', radius=0.05, alpha=0.5)

show(gridplot([[fig1, fig2], [fig3, fig4]], width=300))
```

```python
fig = figure(
    width=400,
    aspect_ratio=1,
    x_axis_label='$$t$$',
    y_axis_label='$$\Omega$$',
)
fig.grid.grid_line_dash = [6, 4]
fig.toolbar.logo = None
fig.toolbar.autohide = True

c = fig.circle(ts1[::20], ys3[::20], color='orange', legend_label='sin + cos', radius=0.05, alpha=0.5)

s1 = Slider(start=0.05, end=1, value=0.05, step=0.05, title="radius", sizing_mode='stretch_width')
s1.js_link('value', c.glyph, 'radius')
s2 = Slider(start=0, end=1, value=0.5, step=0.1, title="alpha", sizing_mode='stretch_width')
s2.js_link('value', c.glyph, 'fill_alpha')

show(row(column(s1, s2, width=200), fig))
```

```python
n = 10000

xs = np.random.standard_normal(n)
ys = np.random.standard_normal(n)

hb = hexbin(xs, ys, 0.2)

fig = figure(width=400, aspect_ratio=1)
fig.grid.visible = False
fig.background_fill_color = all_palettes['Turbo'][256][0]

cmap = linear_cmap('counts', all_palettes['Turbo'][256], 0, max(hb.counts))
fig.hex_tile(size=0.2, source=hb, fill_color=cmap, line_color=None)

s1 = Slider(start=1000, end=100000, value=10000, step=1000, title='N', width=200)

show(column(Div(text='<h1>YOLO</h1>'), row(s1, fig)))
```
